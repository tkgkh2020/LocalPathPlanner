# 自動車の運動と制御 車両運動力学の理論形成と応用 第3章参考 Python

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

### 車両パラメータ設定

```python
m = 1500;   #車両重量 [kg]
I = 2500;   #ヨー軸回りの慣性モーメント [kgm^2]
lf = 1.1;#1.6;   #車両重心から前輪車軸までの距離 [m]
lr = 1.6;#1.1;   #車両重心から後輪車軸までの距離 [m]
l = lf+lr;  #ホイールベース [m]
Kf = 55000;#60000; #前輪のコーナリングフォース [N/rad]
Kr = 60000;#55000; #後輪のコーナリングフォース [N/rad]
A = -m*(lf*Kf-lr*Kr)/(2*l*l*Kf*Kr); #スタビリティファクタ [s^2/m^2]

#シミュレーションのパラメータ設定
dt = 0.001;   #制御周期
tf = 4.0;     #シミュレーション時間
V = 140/3.6;  #車両速度 [m/s]
rho = 15;     #旋回半径 [m]
#delta = 0.04; #タイヤ実舵角 [rad]
```

### 入力の速度

```python
V_vector = np.arange(1,41)
np.shape(V_vector)
```

### タイヤ舵角の関数(旋回半径一定)

```python
def steer_angle(A, rho, l, V):
    steer = (1+A*V*V)*(l/rho)
    return steer
```

### plot

```python
steer_vector = np.array([steer_angle(A, rho, l, V_vector[i]) for i in range(len(V_vector))])
```

```python
np.shape(steer_vector)
```

```python
plt.plot(V_vector, steer_vector)
plt.title("x軸: 速度　y軸: タイヤ舵角", fontname="MS Gothic")
plt.xlabel("x軸: 速度", fontname="MS Gothic")
plt.ylabel("y軸: タイヤ舵角", fontname="MS Gothic")
plt.legend(["タイヤ舵角"], prop={"family":"MS Gothic"})
plt.show()
```

### 旋回半径の関数(タイヤ舵角一定)

```python
steer = 0.04; #タイヤ実舵角 [rad]
def turning_radius(A, steer, l, V):
    rho = (1+A*V*V)*(l/steer)
    return rho
```

### plot

```python
rho_vector = np.array([turning_radius(A, steer, l, V_vector[i]) for i in range(len(V_vector))])
```

```python
plt.plot(V_vector, rho_vector)
plt.title("x軸: 速度　y軸: 旋回半径", fontname="MS Gothic")
plt.xlabel("x軸: 速度", fontname="MS Gothic")
plt.ylabel("y軸: 旋回半径", fontname="MS Gothic")
plt.legend(["旋回半径"], prop={"family":"MS Gothic"})
plt.show()
```

### ヨーレートの関数(タイヤ舵角一定)

```python
steer = 0.04; #タイヤ実舵角 [rad]
def yaw_rate(A, steer, l, V):
    r = (1/(1+A*V*V))*(V/l)*steer
    return r
```

### plot

```python
r_vector = np.array([yaw_rate(A, steer, l, V_vector[i]) for i in range(len(V_vector))])
```

```python
plt.plot(V_vector, r_vector)
plt.title("x軸: 速度　y軸: ヨーレート", fontname="MS Gothic")
plt.xlabel("x軸: 速度", fontname="MS Gothic")
plt.ylabel("y軸: ヨーレート", fontname="MS Gothic")
plt.legend(["ヨーレート"], prop={"family":"MS Gothic"})
plt.show()
```

### 横滑り角の関数(タイヤ舵角一定)

```python
steer = 0.04; #タイヤ実舵角 [rad]
def side_skip(m, A, steer, l, lf, lr, Kr, V):
    b = ((1-(m*lf*V*V)/(2*l*lr*Kr))/(1+A*V*V))*(lr/l)*steer
    return b
```

### plot

```python
b_vector = np.array([side_skip(m, A, steer, l, lf, lr, Kr, V_vector[i]) for i in range(len(V_vector))])
```

```python
plt.plot(V_vector, b_vector)
plt.title("x軸: 速度　y軸: 横滑り角", fontname="MS Gothic")
plt.xlabel("x軸: 速度", fontname="MS Gothic")
plt.ylabel("y軸: 横滑り角", fontname="MS Gothic")
plt.legend(["横滑り角"], prop={"family":"MS Gothic"})
plt.show()
```

### 車両重心点(x座標, y座標)の軌跡

```python
x0 = 0;
y0 = 0;
theta0 = 0;
dt = 1;

theta_vector = []
X_vector = []
Y_vector = []

for t in range(len(V_vector)):
    theta = theta0 + r_vector[t]*dt
    X = x0 + V_vector[t]*np.cos(theta + b_vector[t])*dt
    Y = y0 + V_vector[t]*np.sin(theta + b_vector[t])*dt
    
    X_vector.append(X)
    Y_vector.append(Y)
```

```python
np.shape(Y_vector)
```

### plot

```python
plt.plot(X_vector, Y_vector)
plt.title("車両重心点(x座標、y座標)", fontname="MS Gothic")
plt.xlabel("x軸: x座標", fontname="MS Gothic")
plt.ylabel("y軸: y座標", fontname="MS Gothic")
#plt.legend(["車両重心点(x座標、y座標)"], prop={"family":"MS Gothic"})
plt.show()
```
